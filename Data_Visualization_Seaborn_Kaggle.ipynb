```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

```python
fifa_data = pd.read_csv('fifa.csv', index_col='Date')
fifa_data.head()
```

```python
list(fifa_data.columns)
```

```python
plt.figure(figsize = (12,6))
plt.title('FIFA Plot for ranking of Important country')
sns.lineplot(data = fifa_data)
```

```python
sns.lineplot(data = fifa_data['ARG'], label = 'Argantinia')
sns.lineplot(data = fifa_data['ESP'], label = 'Espania')
plt.xlabel('Date')
```

```python
fifa_5 = fifa_data.head(5)
sns.barplot(x = fifa_5.index, y= fifa_5['ARG'])
plt.figure(figsize = (10,6)) #why this is not excuted
plt.ylabel('Argentina')
plt.title('FIFA Report for Argentina')
```

```python
plt.figure(figsize=(12,6))
plt.title('FIFA report for Espain')
sns.heatmap(data = fifa_5, annot = True)
#the brighter, The better 
```
